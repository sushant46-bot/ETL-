```python
dict1 = {"emp":[1,2,3],"depno":[10,20,30]}
dict2 = {"emp":[4,5,6,8,9],"depno":[70,80,90,100,110]}
```

```python
df1 = pd.DataFrame(dict1)
df2 = pd.DataFrame(dict2)
```

```python
#concat by default axis = 0   (one above other) and for axis = 1 (side by side from two dict data)
df_ans = pd.concat([df1,df2],axis=1)
```

```python
df_ans
```

```python
#joins this is not SL joins this is just the union, intersect 
df_ans = pd.concat([df1,df2],axis=0,join = 'inner')
```

```python
df_ans
```
