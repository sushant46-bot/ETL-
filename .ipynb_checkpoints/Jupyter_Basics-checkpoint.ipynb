```python
#read two diffrnt csv file to make join
df_emp = pd.read_csv(r"C:\Users\user\PycharmProjects\PythonProject_01\Pandas basics\employees.csv")
df_dept = pd.read_csv(r"C:\Users\user\PycharmProjects\PythonProject_01\Pandas basics\Department.csv")
```

```python
df_dept
```

```python
# join the table with inner join 
df_emp.merge(df_dept,on = 'DeptNo', how = 'inner')
```

```python
# join the table with right join 
df_emp.merge(df_dept,on = 'DeptNo', how = 'right')
```
